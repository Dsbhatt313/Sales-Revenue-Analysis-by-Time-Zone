```python
import pandas as pd
import pytz
from datetime import datetime
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("Sales-products-tz-mod.csv", index_col = 'SaleID')
df_sales_data = data.copy()
df_sales_data
```

```python
df_sales_data['MOS-UTC'] = [pd.to_datetime(df_sales_data['MOS'].iloc[i]).tz_localize(df_sales_data['TimeZone'].iloc[i]).tz_convert('UTC') for i in range(len(df_sales_data))]
df_sales_data.head()
```

```python
cols_name_reordered = ['RetailerCountry', 'MOS-UTC', 'MOS', 'TimeZone', 'OffsetUTC', 'RetailerType', 'Product', 'Sales Revenue ($)']
df_sales_data = df_sales_data[cols_name_reordered]
df_sales_data.head()
```

```python
df_sales_data_viz = df_sales_data.reset_index()
df_sales_data_viz
```

```python
df_sales_data_viz = df_sales_data_viz.set_index('MOS-UTC').sort_index()
df_sales_data_viz.head()
```

```python
plt.scatter(df_sales_data_viz['OffsetUTC'], df_sales_data_viz['RetailerType'])
plt.show()
```

```python
df_sales_data_viz.groupby('TimeZone').sum(numeric_only = True)
```

```python
total_sales_by_timezones = df_sales_data_viz.groupby('TimeZone').sum(numeric_only = True)
total_sales_by_timezones
```

```python
plt.bar(total_sales_by_timezones.index, height = total_sales_by_timezones['Sales Revenue ($)'])
plt.show()
```

```python
plt.figure(figsize = (7, 6))

plt.bar(total_sales_by_timezones.index, 
        height = total_sales_by_timezones['Sales Revenue ($)'], 
        color = '#108A99')

plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title('Total Sales Revenue by Time Zone', 
          fontsize = 18, 
          fontweight = 'bold')
plt.ylabel('Sales Revenue ($)', fontsize = 13)

plt.show()
```

```python
df_sales_data_viz.groupby('TimeZone').mean(numeric_only = True)
```

```python
df_sales_data_viz.groupby('TimeZone').mean(numeric_only = True).sort_values('Sales Revenue ($)')
```

```python
avg_by_timezones = df_sales_data_viz.groupby('TimeZone').mean(numeric_only = True).sort_values('Sales Revenue ($)', ascending = False)
avg_by_timezones
```

```python
plt.bar(avg_by_timezones.index, height = avg_by_timezones['Sales Revenue ($)'])
```

```python
plt.figure(figsize = (7, 6))

plt.bar(avg_by_timezones.index, height = avg_by_timezones['Sales Revenue ($)'], color = ['#2358aa', '#85aeee', '#416b90', '#5699c6', '#75b9ea', 'lightblue'])

plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title('Average Sales Revenue by Time Zone', 
          fontsize = 18, 
          fontweight = 'bold')
plt.ylabel('Average price ($)', fontsize = 13)

plt.show()
```
